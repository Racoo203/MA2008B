```python
import numpy as np
import sympy as sp
```

```python
# Define a symbolic function
x, y = sp.symbols('x y')
f = 4*x + 7*y + x*y - x**2 - 2*y**2

# Compute symbolic gradients
grad_f = [sp.diff(f, var) for var in (x, y)]

# Convert symbolic gradients to numerical functions
grad_f_num = [sp.lambdify((x, y), g, 'numpy') for g in grad_f]

# Gradient ascent parameters
learning_rate = 0.1
point = np.array([2.0, 2.0])  # Initial guess

epsilon = np.inf
iteration = 0
# Gradient ascent loop
while epsilon > 0.01:
    iteration += 1
    grad = np.array([g(point[0], point[1]) for g in grad_f_num]).round(3)
    epsilon = np.linalg.norm(grad).round(3)
    point += np.round(learning_rate * grad, 3)
    
    print(f"Iteration {iteration}")
    print(f"Grad is: {grad}")
    print(f"Grad norm is: {epsilon}")
    print(f"Maximum at: {point}")
    print("="*40)
```
